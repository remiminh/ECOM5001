Calculate expected return of Ferrari stock on the dates after race dates using market model from 2016 to 2022

```python
# merge the market index and Ferrari stock dataframes by year
columns_to_merge = ['Date', 'Daily return']
stock_return_2016 = pd.merge(ferrari_stock_2016_df[columns_to_merge], nyse_2016_df[columns_to_merge], on='Date').rename(columns={'Daily return_x': 'Stock return', 'Daily return_y': 'Market return'})
stock_return_2017 = pd.merge(ferrari_stock_2017_df[columns_to_merge], nyse_2017_df[columns_to_merge], on='Date').rename(columns={'Daily return_x': 'Stock return', 'Daily return_y': 'Market return'})
stock_return_2018 = pd.merge(ferrari_stock_2018_df[columns_to_merge], nyse_2018_df[columns_to_merge], on='Date').rename(columns={'Daily return_x': 'Stock return', 'Daily return_y': 'Market return'})
stock_return_2019 = pd.merge(ferrari_stock_2019_df[columns_to_merge], nyse_2019_df[columns_to_merge], on='Date').rename(columns={'Daily return_x': 'Stock return', 'Daily return_y': 'Market return'})
stock_return_2020 = pd.merge(ferrari_stock_2020_df[columns_to_merge], nyse_2020_df[columns_to_merge], on='Date').rename(columns={'Daily return_x': 'Stock return', 'Daily return_y': 'Market return'})
stock_return_2021 = pd.merge(ferrari_stock_2021_df[columns_to_merge], nyse_2021_df[columns_to_merge], on='Date').rename(columns={'Daily return_x': 'Stock return', 'Daily return_y': 'Market return'})
stock_return_2022 = pd.merge(ferrari_stock_2022_df[columns_to_merge], nyse_2022_df[columns_to_merge], on='Date').rename(columns={'Daily return_x': 'Stock return', 'Daily return_y': 'Market return'})
```

```python
# Calculate expected return of Ferrari stock using market model: alpha + beta * Market return
stock_return_2016['Expected_return'] = alpha_2016 + beta_2016 * stock_return_2016['Market return']
stock_return_2017['Expected_return'] = alpha_2017 + beta_2017 * stock_return_2017['Market return']
stock_return_2018['Expected_return'] = alpha_2018 + beta_2018 * stock_return_2018['Market return']
stock_return_2019['Expected_return'] = alpha_2019 + beta_2019 * stock_return_2019['Market return']
stock_return_2020['Expected_return'] = alpha_2020 + beta_2020 * stock_return_2020['Market return']
stock_return_2021['Expected_return'] = alpha_2021 + beta_2021 * stock_return_2021['Market return']
stock_return_2022['Expected_return'] = alpha_2022 + beta_2022 * stock_return_2022['Market return']
```

Calculate abnormal return of Ferrari stock: actual return - expected return

```python
stock_return_2016['Abnormal_return'] = stock_return_2016['Stock return'] - stock_return_2016['Expected_return'] 
stock_return_2017['Abnormal_return'] = stock_return_2017['Stock return'] - stock_return_2017['Expected_return'] 
stock_return_2018['Abnormal_return'] = stock_return_2018['Stock return'] - stock_return_2018['Expected_return'] 
stock_return_2019['Abnormal_return'] = stock_return_2019['Stock return'] - stock_return_2019['Expected_return'] 
stock_return_2020['Abnormal_return'] = stock_return_2020['Stock return'] - stock_return_2020['Expected_return'] 
stock_return_2021['Abnormal_return'] = stock_return_2021['Stock return'] - stock_return_2021['Expected_return'] 
stock_return_2022['Abnormal_return'] = stock_return_2022['Stock return'] - stock_return_2022['Expected_return'] 
```

```python
# combine the datasets from 2016 to 2022
stock_return_df = pd.concat([stock_return_2016, 
                             stock_return_2017, 
                             stock_return_2018, 
                             stock_return_2019,
                             stock_return_2020,
                             stock_return_2021,
                             stock_return_2022], ignore_index=True)
stock_return_df
```

Check the race events following with a closed trading date in stock market.

Calculate the average abnormal return from 2016 to 2022

```python
AAR = stock_return_df['Abnormal_return'].mean()
AAR
```

Calculate the sum of squared residuals

```python
sum_squared_residuals = sum(residuals_2016**2) + sum(residuals_2017**2) + sum(residuals_2018**2) + sum(residuals_2019**2) + sum(residuals_2020**2) + sum(residuals_2021**2) + sum(residuals_2022**2)
sum_squared_residuals                        
```

```python
# total number of observations
N = len(residuals_2016) + len(residuals_2017) + len(residuals_2018) + len(residuals_2019) + len(residuals_2020) + len(residuals_2021) + len(residuals_2022)
N
```

Calculate the variance for the average abnormal returns

```python
var_AAR = sum_squared_residuals / N**2
var_AAR
```

CaCalculate the test statistic (𝜃1)

```python
import numpy as np
test_statistic = AAR / np.sqrt(var_AAR)
test_statistic
```

Calculate p-value

```python
import scipy.stats as stats

degrees_of_freedom = N - 1  # N is the number of observations

p_value = 2 * (1 - stats.t.cdf(np.abs(test_statistic), degrees_of_freedom))
p_value
```
